```python
# Importando as bibliotecas que cont√©m os comandos

import pandas as pd
import numpy as np
import io
import folium

from google.colab import files

import plotly.express as px
```

```python
# Carregando um arquivo do meu pc para a nuvem

file_uploaded = files.upload()
```

```python
# Comando para carregar o arquivo da Nuvem para a IDE do Google Collabs

df = pd.read_csv( io.BytesIO( file_uploaded['AB_NYC_2019.csv'] ) )
```

```python
# Removendo os N#A da planilha/ df1 ser√° a minha planilha - os dados v√°zios (N#A)
df1 = df.dropna()


df1 = df1.loc[df1[ 'availability_365' ] > 0, :] # Remover im√≥veis com a coluna availability_365 igual a 0 (N√£o disponivel para loca√ß√£o)
```

```python
# 1. Qual o aluguel (di√°ria) mais caro por regi√£o da cidade de NY, apenas os im√≥veis dispon√≠veis para alugar?
      # R: A di√°ria mais cara da regi√µ do Bronx √© de U$ 800, Brooklyn √© de U$ 8000, Manhattan √© de U$ 10000, Queens √© de U$ 2600 e por fim de Stand Islands √© de U$ 625

cols = [ 'neighbourhood_group', 'price' ]

data_plot = df1.loc[:, cols].groupby( 'neighbourhood_group' ).max().reset_index()

px.bar(
    data_plot,                ###### Onde est√£o os dados
    x='neighbourhood_group',  ###### Eixo X = Horizontal
    y='price'                 ###### Eixo Y = Vertical

)
```

```python
from folium.map import Popup
# 2. Onde est√£o localizados os im√≥veis com o valor do aluguel mais caro na cidade de NY?

# O conjunto de dados com a coordenadas
cols = ['price', 'neighbourhood_group', 'latitude', 'longitude']

data_plot = df1.loc[:,cols].groupby('neighbourhood_group').max().reset_index()

# Um comando para desenhar um mapa
map = folium.Map()

# Colocar os pontos de latitude e longitude dentro do mapa

for index, location_info in data_plot.iterrows():
  folium.Marker( [location_info['latitude'], location_info['longitude']],
              popup=location_info['neighbourhood_group']).add_to( map )

map
```

```python
# 3. Conseguimos saber onde est√£o localizados os im√≥veis por seu tipo?

cols = ['neighbourhood_group', 'room_type', 'latitude', 'longitude']
data_plot = df1.loc[:,cols].sample ( 100 ) # SAMPLE escolhe 100 linhas aleat√≥rias

#Criar uma nova coluna

data_plot.loc[:, 'color'] = 'NA' #Qnd quero atribuir √© um unico IGUAL

#Qnd quero comparar s√£o 2 IGUAIS

data_plot.loc[data_plot['room_type'] == 'Private room', 'color'] = 'darkgreen'
data_plot.loc[data_plot['room_type'] == 'Entire home/apt', 'color'] = 'darkred'
data_plot.loc[data_plot['room_type'] == 'Shared room', 'color'] = 'purple'

# Desenhar o mapa e colocar os pontos de latitude e longitude

map = folium.Map()

for index, location_info in data_plot.iterrows():
  folium.Marker( [location_info['latitude'], location_info['longitude']],
              popup=location_info['neighbourhood_group'],
              icon=folium.Icon( color=location_info['color'] ) ).add_to( map )

map
```
